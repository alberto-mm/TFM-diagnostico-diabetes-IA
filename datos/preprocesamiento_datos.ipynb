__Todos los atributos aparecen como tipo _object_ por lo que hay que transformarlos a float.__

Para ello, primero hay que eliminar la primera fila que indica el grupo al que pertenecen las mediciones de esa columna.

Si se observa el dataset, esto no es necesario ya que los t√≠tulos de las columnas son distintos para cada grupo. Los t√≠tulos que incluyen un "_.1_" pertenecen al grupo 1 de diab√©ticos (sin √∫lceras visibles), el "_.2_" hace referencia al grupo 2 de diab√©ticos (con √∫lceras visibles) y el resto al grupo de control.

La informaci√≥n almacenada en cada columna es la siguiente:

- AF365: Amplitud de fluorescencia 365nm.
- AF450: Amplitude de fluorescencia 450nm.
- 35 deg: Perfusi√≥n de sangre a 35 grados Celsius.
- 42 deg: Perfusi√≥n de sangre a 42 grados Celsius.

```python
df_completo = df_completo.drop(0)
```

```python
df_completo
```

```python
# Es necesario reindexar para que la primera fila vuelva a ser la 0
df_completo = df_completo.reset_index(drop=True)
```

```python
df_completo
```

---

Es necesario hacer una conversi√≥n previa de los valores ya que los decimales est√°n separados por una coma (,) y no por un punto (.). Esto puede provocar un error al usar el m√©todo *astype(float)*.

```python
df_completo = df_completo.apply(lambda x: x.str.replace(',','.'))
```

```python
# Se convierte el dataset a tipo float
df_completo = df_completo.astype(float)
```

### One-hot encode

```python
datos_onehot = datos.copy()

datos_onehot['Grupo'] = datos_onehot['Grupo'].replace(0, 'Control')
datos_onehot['Grupo'] = datos_onehot['Grupo'].replace(1, 'Diabetes 1')
datos_onehot['Grupo'] = datos_onehot['Grupo'].replace(2, 'Diabetes 2')


datos_onehot
```

```python
# One-hot encoding a la columna de grupo
one_hot = pd.get_dummies(datos['Grupo'])

# Se elimina la columna "Grupo"
datos_onehot = datos_onehot.drop('Grupo', axis=1)

# Se una de dataframe con el one-hot
datos_onehot = datos_onehot.join(one_hot)

datos_onehot
```

```python
# Cambiar el nombre de las columnas

datos_onehot.columns = ['AF365', 'AF450', '35 deg', '42 deg', 'Control', 'Diabetes 1', 'Diabetes 2']

datos_onehot
```

```python
# Guardar dataframe
datos_onehot.to_csv('datos_onehot.csv', header=True, index=False)
```
